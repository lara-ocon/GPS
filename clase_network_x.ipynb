```python
import networkx as nx
import matplotlib.pyplot as plt
```

Grafos no dirigidos

```python
G = nx.Graph()
""" forma 1
for i in range(1,6):
    G.add_node(i)
"""
G.add_nodes_from(range(1,6)) # añade los nodos de una lista
G.add_edges_from([(1,2), (1,4), (3,4), (2,4), (3,2), (3,5)])
print(G)
print(G.nodes())
print(G.edges())
```

eliminar nodos (igual que añadir)

```python
G.remove_node(5)
print(G)
print(G.nodes())
print(G.edges())
print("ha pasado de 6 a 5 aristas luego tamb borra las aristas del nodo 5")
print("Parece que da error si el nodo no esta en el grafo")
```
