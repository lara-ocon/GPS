```python
import pandas as pd
import regex as re
```

```python
cruces = pd.read_csv('cruces.csv', sep=';', encoding='latin-1')
cruces.head()
```

```python
direccion = input('Introduce la dirección; ')
calle, numero = direccion.split(',')

# vamos a buscas el primer cruce con dicha calle
encontrado = 'No hay ninguna'
for fila in cruces.itertuples():
    if re.search(fila[5],calle, re.I):
        encontrado = fila[2]
        break

print(encontrado)
```
