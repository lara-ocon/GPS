```python
import pandas as pd
```

```python
direcciones = pd.read_csv('direcciones.csv', sep=';', encoding='latin-1', low_memory=False)
direcciones = direcciones.sort_values(by=['Codigo de via'] )
```

```python
df_autovias = direcciones[direcciones['Clase de la via'].str.contains('AUTO')]
df_autovias.drop(['Tipo de la via a la que pertenece el numero'], axis=1, inplace=True)
df_autovias.drop(['Codigo de barrio'], axis=1, inplace=True)
df_autovias.drop(['Partï¿½cula de la vï¿½a'], axis=1, inplace=True)
df_autovias.drop(['Seccion de carteria'], axis=1, inplace=True)
df_autovias.drop(['Seccion censal'], axis=1, inplace=True)
df_autovias.drop(['Zona Servicio Estacionamiento Regulado'], axis=1, inplace=True)
df_autovias.drop(['Categoria fiscal'], axis=1, inplace=True)
df_autovias.drop(['Parcela catastral del numero'], axis=1, inplace=True)
df_autovias.drop(['Tipo de denominacion de la via'], axis=1, inplace=True)
df_autovias.drop(['Zona de valor del numero'], axis=1, inplace=True)
df_autovias.drop(['Tipologia del numero'], axis=1, inplace=True)
df_autovias.head(20)
```

```python
df_autovias.to_csv('autovias.csv', sep=';', encoding='latin-1')
```
